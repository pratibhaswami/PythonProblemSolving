20. Finding the maximum length word in a list:

words = ["apple", "banana", "cherry", "date"]

```python
words = ["apple", "banana", "cherry", "date"]

max_len= 0
for i in words:
    if len(i)>max_len:
        
        max_len+=len(i)
        print(i)
```
